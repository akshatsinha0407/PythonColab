```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = fetch_california_housing()
```

```python
df
```

```python
dataset = pd.DataFrame(df.data)
```

```python
dataset
```

```python
dataset.columns = df.feature_names
```

```python
dataset.head()
```

```python
X =dataset
y = df.target
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.30, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
X_train
```

```python
X_train = sc.fit_transform(X_train)
X_train
```

```python
## all though our data vas allredy standerdise
```

```python
X_test
```

```python
X_test = sc.transform(X_test)
X_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import cross_val_score
```

```python
reg = LinearRegression()
reg.fit(X_train,y_train)
```

```python
mse = cross_val_score(reg,X_train,y_train,scoring='neg_mean_squared_error',cv=5)
```

```python
np.mean(mse)
```

```python
reg_pred = reg.predict(X_test)
```

```python
reg_pred
```

```python
import seaborn as sns
residuals = reg_pred - y_test

sns.kdeplot(residuals, shade=True)  # KDE plot with shaded area
plt.xlabel("Residuals")
plt.ylabel("Density")
plt.title("Residuals Distribution")
plt.show()
```
